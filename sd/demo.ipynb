```python
import model_loader
import pipeline
from PIL import Image
from transformers import CLIPTokenizer
from torch

DEVICE = "cpu"
ALLOW_CUDA = False
ALLOW_MPS = False
if torch.cuda.is_available() and ALLOW_CUDA:
    DEVICE ="cuda"
elif(torch.has_mps or torch.backend.mps.is_available()) and ALLOW_MPS:
    DEVICE = "mps"
print(f"using device {DEVICE}")
tokenizer = CLIPTokenizer("../data/vocab.json",merges_file = "..data/merges.txt")
model_file = "../data/v1-5-pruned-emaonly.ckpt"
models = model_loader.preload_models_from_standard_weights(model_file,DEVICE)

#going from text to image:

prompt = "A potato dressed like a trained astronaut on the moon exploring the deep space, a cartoon style illustration,highly detailed."
uncond_prompt = "" # also used as negative prompt
do_cfg = True
cfg_scale = 7

#going from image to image
input_image =None
image_path = '../images/Potato.jpg'
input_image = Image.open(image_path)
strength = 0.9

sampler = "ddpm"
num_inference_steps = 50
seed = 42

output_image = pipeline.generate(
    prompt = prompt,
    uncond_prompt = uncond_prompt,
    input_image = input_image,
    strength = strength,
    cfg_scale = cfg_scale,
    sampler_name = sampler,
    num_inference_steps = num_inference_steps,
    seed = seed,
    models = models,
    device = DEVICE
    idle_device = "cpu",
    tokenizer = tokenizer
)

Image.from_array(output_image)
```
